```
import pandas as pd
from sklearn import datasets
from matplotlib import pyplot
from matplotlib import pyplot
from matplotlib import pylab
import numpy as np
from sklearn.decomposition import PCA
from sklearn.decomposition import NMF
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```
data = datasets.load_breast_cancer()
```

```
cancer_df = pd.DataFrame(data=np.c_[data['data'], data['target']],columns= list(data['feature_names']) + ['target'])
```

```
cancer_df.head()
```

```
pca = PCA(n_components=3, whiten=True)
```

```
decomp_data=pca.fit_transform(cancer_df.drop('target',axis=1))
```

```
pca.explained_variance_ratio_
```

```
pyplot.scatter(x=decomp_data[::,0],y=decomp_data[::,1],c=cancer_df['target'])
```

```
pyplot.scatter(x=decomp_data[::,0],y=decomp_data[::,2],c=cancer_df['target'])
```

```
pca.components_
```

```
pyplot.matshow(pca.components_,cmap='viridis')
pyplot.yticks([0,1,2],["1st",'2nd','3rd'])
pyplot.colorbar()
pyplot.xticks(range(31),cancer_df.drop('target',axis=1).columns,
              rotation=45,ha='left')
pyplot.xlabel('feature')
pyplot.ylabel('PC')
```

```
pyplot.scatter(x=data_nmf[::,0],y=data_nmf[::,2],c=cancer_df['target'])
```

```
pyplot.scatter(x=data_nmf[::,1],y=data_nmf[::,2],c=cancer_df['target'])
```

```
iris_data = datasets.load_iris()
iris_df = pd.DataFrame(data=np.c_[iris_data['data'], iris_data['target']],columns= list(iris_data['feature_names']) + ['target'])
```

```
pca = PCA(n_components=2, whiten=True)
decomp_data=pca.fit_transform(iris_df.drop('target',axis=1))
decomp_data
```

```
pyplot.scatter(x=decomp_data[::,0],y=decomp_data[::,1],c=iris_df['target'])
```

```
pyplot.matshow(pca.components_,cmap='viridis')
pyplot.yticks([0,1],["1st",'2nd'])
pyplot.colorbar()
pyplot.xticks(range(4),iris_df.drop('target',axis=1).columns,
              rotation=60,ha='left')
pyplot.xlabel('feature')
pyplot.ylabel('PC')
```

```
nmf = NMF(n_components=2)
data_nmf = nmf.fit_transform( iris_df.drop('target',axis=1) )
```

```
pyplot.scatter(x=data_nmf[::,0],y=data_nmf[::,1],c=iris_df['target'],cmap='plasma')
```

```
pyplot.matshow(nmf.components_,cmap='viridis')
pyplot.yticks([0,1],["1st",'2nd'])
pyplot.colorbar()
pyplot.xticks(range(4),iris_df.drop('target',axis=1).columns,
              rotation=60,ha='left')
pyplot.xlabel('feature')
pyplot.ylabel('PC')
```
